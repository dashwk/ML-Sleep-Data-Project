## Building The Pipeline

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

log_reg_pipeline = make_pipeline(
    StandardScaler(),
    LogisticRegression()
)
```

Right now, log_reg_pipeline is just a definition. When we call X_train, y_train on it, it will 
- Take X_train, fit the StandardScaler on it (compute each feature's mean and standard deviation from the training data), and transform X_train into scaled values
- Take that scaled version of X_train and feed it into LogisticRegression().fit() along with y_train

Without a pipeline, the approach is to scale the entire dataset once before splitting into train/test (which is wrong). That's a subtle bug: the scaler would compute mean/std using information from the test set too, meaning your "unseen" test data quietly influenced how the training data got transformed. 

Pipelines force the correct order - fit the scaler only on training data, then when we call it on the test data it reuses the training set's mean/std. 

Bottom line: **the model should never learn anything, even indirectly, from data it's supposed to be evaluated on.**

```python
log_reg_pipeline.fit(X_train, y_train)
```

```python
log_reg_pipeline.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(log_reg_pipeline, X, y, cv=5, scoring='accuracy')
print(scores, scores.mean())
```

4 perfect folds, 1 dropped to 77%

```python
from sklearn.metrics import confusion_matrix, classification_report
y_pred = log_reg_pipeline.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

For the 15, 0 and 1,6:
all 15 "Good" cases were correctly predicted as Good (no false positives). Of 7 "Poor" cases, 6 were caught correctly and 1 was missed — predicted as Good when it was actually Poor. 

Recall on "Poor" is 0.86 — lower than everything else — which is the number that captures this directly. Worth a line in your writeup: "the model's weakest point is catching poor sleep cases; missing one recall point matters more here than a missed 'Good' case would."

### GridSearchCV

Evaluating each combination of hyperparameter values and returning which one ran best

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'logisticregression__C': [0.01, 0.1, 1, 10, 100]}

grid_search = GridSearchCV(log_reg_pipeline, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
print(grid_search.best_score_)
```

C=1 is the default: the default hyperparameter was already close to optimal for this data. For writeup: "tested C values from 0.01 to 100 via GridSearchCV; the default (C=1) was already optimal, suggesting the classes are cleanly separable enough that regularization strength isn't the limiting factor here."

```python
final_log_reg = grid_search.best_estimator_
final_score = final_log_reg.score(X_test, y_test)
print(final_score)
```
